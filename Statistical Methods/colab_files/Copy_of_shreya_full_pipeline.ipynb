```python
def convert_pfm_to_ppm(pfm):
  for i in range(pfm.shape[1]):
    sum = 0
    for j in range(pfm.shape[0]):
      sum+= pfm[j][i] 
    for j in range(pfm.shape[0]):
      prob = float(pfm[j][i]/sum)
      ppm[j][i] = float(prob)
  return(ppm)
  
```

```python
p1=['A','C','G','T']
str1=''
def ppm_to_motif(ppm):
  motif = []
  mot=[]
  for i in range(ppm.shape[1]):
    arr = []
    for j in range(ppm.shape[0]):
      arr.append(float(ppm[j][i]))
    choice = np.random.choice(p1,p=arr) #motif is generated here
    mot.append(choice)
  
  return (''.join(mot))
  #return choice
```

```python
from sklearn.metrics import roc_auc_score
from sklearn.metrics import roc_curve, auc
from sklearn.metrics import accuracy_score
import pandas as pd 

var=4
def markov_model_motif(df1):
    """
    for a second degree motif markov model 
    returns dict2, dict3: count of the appearance of eg: A->C and eg: AC->T occuring
    """
    dict2, dict3 = {}, {}
    for j in range(len(df1)):
      seq = df1[j]
      for i in range(len(seq)-var): #dict2
        if seq[i:i+var] not in dict2: 
          dict2[seq[i:i+var]] = 1
        else: 
          dict2[seq[i:i+var]] += 1

      for i in range(len(seq)): #dict3
        if seq[i:i+var+1] not in dict3: 
          dict3[seq[i:i+var+1]] = 1
        else: 
          dict3[seq[i:i+var+1]] += 1
      
    return dict2, dict3


def motif_score(dict2, dict3, seq):
    """
    returns the score of th markov model motif 
    """
    s = seq
    import math

    score = 1
    for i in range(0, len(s)):
        if s[i : i + var+1] in dict3:
            score = score * dict3[s[i : i + var+1]]
    return math.log(score)


def train():
    # df1 = df['seq']
    seq_list = df1.to_list()  # list of sequences
    dict2, dict3 = markov_model_motif(df1)  # markov model for the motif.txt file
    # print("####################################################")
    
    dict2_non, dict3_non = markov_model_motif(
        df3
    )  # markov model for the non motif file
    # now, testing each sequence and predicting their class using the 2 dictionaries

    for i in range(len(df4)):
        list1 = []
        list2 = []
    for i in range(len(df4)):
        seq = df4["seq"][i]
        """
        1. score it using ddict3
        2. score it using dict3_non 
        3. whichever score higher, give value of that class 
        4. add seq, class to a list, append that list to the df 
        5. write this file 
        """

        score_motif = motif_score(dict2, dict3, seq)
        score_nonmotif = motif_score(dict2_non, dict3_non, seq)
        list1.append(score_motif-score_nonmotif)
        if score_motif > score_nonmotif:
            list2.append(1)
        else:
            list2.append(0)
    df4["predicted"] = list2
    df4["log odd score"] = list1
    print(df4)
    

def accuracy():
    y_true = df4["Bound"]
    y_pred = df4["predicted"]
    return accuracy_score(y_true, y_pred)

colnames2 = ["FoldID", "EventID", "start_index", "seq", "Bound"]
colnames = ["FoldID", "EventID", "seq", "Bound"]
df = pd.read_csv("/content/drive/MyDrive/BTech Proj/Data/Base_NonMotif_Files/PFM_Motif-1_Human.txt", delimiter="\t", names=colnames2, header=None)

df2 = pd.read_csv("/content/drive/MyDrive/BTech Proj/Data/Base_NonMotif_Files/train/1000_314_non-motif-1.txt", delimiter="\t", names=colnames, header=None)
df1 = df["seq"]
df3 = df2["seq"]
df1.reset_index(drop=True, inplace=True)
df3.reset_index(drop=True, inplace=True)
df4 = pd.read_csv("/content/drive/MyDrive/BTech Proj/Data/Base_NonMotif_Files/test/2000_314_mixed_Human.txt", delimiter="\t", names=colnames2, header=None)

df4=df4.rename(columns=df4.iloc[0]).drop(df4.index[0])
df4['Bound']=df4['Bound'].astype('int32')
df4.reset_index(drop=True, inplace=True)
train()
acc = accuracy()
print(acc)
# dict2, dict3 = markov_model_motif(df1)
# score = motif_score(dict2, dict3)
```
