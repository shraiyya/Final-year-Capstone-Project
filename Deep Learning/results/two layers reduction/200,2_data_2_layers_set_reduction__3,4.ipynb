```python
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import Flatten, Dropout
from keras.layers.convolutional import Conv1D
from keras.layers.convolutional import MaxPooling1D
from keras.layers.embeddings import Embedding
from keras.preprocessing import sequence
from keras.layers import BatchNormalization
from keras.layers import InputLayer, LeakyReLU

model1 = Sequential(InputLayer(batch_input_shape= (None, 200, 4),
    dtype= 'float32',
    ragged= False,
    sparse= False))

model1.add(Conv1D(activation= 'relu',
    activity_regularizer= None,
    batch_input_shape= (None, 200, 4),
    bias_constraint= None,
    bias_initializer= {'class_name': 'Zeros', 'config': {}},
    bias_regularizer= None,
    data_format= 'channels_last',
    dilation_rate= (1,),
    dtype= 'float32',
    filters= 480,
    groups= 1,
    kernel_constraint= {'class_name': 'MaxNorm',
     'config': {'axis': 0, 'max_value': 1}},
    kernel_initializer= {'class_name': 'VarianceScaling',
     'config': {'distribution': 'uniform',
      'mode': 'fan_avg',
      'scale': 1.0,
      'seed': None}},
    kernel_regularizer= {'class_name': 'L1L2',
     'config': {'l1': 9.99999993922529e-09, 'l2': 4.999999987376214e-07}},
    kernel_size= (9,),
    padding= 'valid',
    strides= (1,),
    trainable= True,
    use_bias= True))

model1.add(BatchNormalization(
    beta_constraint= None,
    beta_initializer= {'class_name': 'Zeros', 'config': {}},
    beta_regularizer= None,
    center= True,
    dtype= 'float32',
    epsilon= 0.001,
    gamma_constraint= None,
    gamma_initializer= {'class_name': 'Ones', 'config': {}},
    gamma_regularizer= None,
    momentum= 0.99,
    moving_mean_initializer= {'class_name': 'Zeros', 'config': {}},
    moving_variance_initializer= {'class_name': 'Ones', 'config': {}},
    scale= True,
    trainable=True))

model1.add(LeakyReLU(alpha= 0.0,
    dtype='float32',
    trainable= True))

model1.add(MaxPooling1D(data_format= 'channels_last',
    dtype= 'float32',
    padding='valid',
    pool_size=(9,),
    strides= (3,),
    trainable= True))

model1.add(Dropout(dtype= 'float32',
    noise_shape= None,
    rate= 0.2,
    seed= None,
    trainable= True))

model1.add(Conv1D(activation= 'relu',
    activity_regularizer= None,
    bias_constraint= None,
    bias_initializer= {'class_name': 'Zeros', 'config': {}},
    bias_regularizer= None,
    data_format= 'channels_last',
    dilation_rate= (1,),
    dtype= 'float32',
    filters= 480,
    groups= 1,
    kernel_constraint= {'class_name': 'MaxNorm',
     'config': {'axis': 0, 'max_value': 0.9}},
    kernel_initializer= {'class_name': 'VarianceScaling',
     'config': {'distribution': 'uniform',
      'mode': 'fan_avg',
      'scale': 1.0,
      'seed': None}},
    kernel_regularizer= {'class_name': 'L1L2',
     'config': {'l1': 9.99999993922529e-09, 'l2': 4.999999987376214e-07}},
    kernel_size= (4,),
    padding= 'valid',
    strides= (1,),
    trainable= True,
    use_bias= True))

model1.add(BatchNormalization(beta_constraint= None,
    beta_initializer= {'class_name': 'Zeros', 'config': {}},
    beta_regularizer= None,
    center= True,
    dtype= 'float32',
    epsilon= 0.001,
    gamma_constraint= None,
    gamma_initializer= {'class_name': 'Ones', 'config': {}},
    gamma_regularizer= None,
    momentum= 0.99,
    moving_mean_initializer= {'class_name': 'Zeros', 'config': {}},
    moving_variance_initializer= {'class_name': 'Ones', 'config': {}},
    scale= True,
    trainable=True))

model1.add(LeakyReLU(alpha= 0.0,
    dtype='float32',
    trainable= True))

model1.add(MaxPooling1D(data_format='channels_last',
    dtype='float32',
    padding='valid',
    pool_size= (4,),
    strides=(2,),
    trainable=True))

model1.add(Dropout(dtype= 'float32',
    noise_shape= None,
    rate= 0.2,
    seed= None,
    trainable= True))

###

###
model1.add(Conv1D(activation= 'relu',
    activity_regularizer= None,
    bias_constraint= None,
    bias_initializer= {'class_name': 'Zeros', 'config': {}},
    bias_regularizer= None,
    data_format= 'channels_last',
    dilation_rate= (1,),
    dtype= 'float32',
    filters= 320,
    groups= 1,
    kernel_constraint= {'class_name': 'MaxNorm',
     'config': {'axis': 0, 'max_value': 1}},
    kernel_initializer= {'class_name': 'VarianceScaling',
     'config': {'distribution': 'uniform',
      'mode': 'fan_avg',
      'scale': 1.0,
      'seed': None}},
    kernel_regularizer= {'class_name': 'L1L2',
     'config': {'l1': 9.99999993922529e-09, 'l2': 4.999999987376214e-07}},
    kernel_size= (1,),
    padding= 'valid',
    strides= (1,),
    trainable= True,
    use_bias= True))

model1.add(BatchNormalization(beta_constraint= None,
    beta_initializer= {'class_name': 'Zeros', 'config': {}},
    beta_regularizer= None,
    center= True,
    dtype= 'float32',
    epsilon= 0.001,
    gamma_constraint= None,
    gamma_initializer= {'class_name': 'Ones', 'config': {}},
    gamma_regularizer= None,
    momentum= 0.99,
    moving_mean_initializer= {'class_name': 'Zeros', 'config': {}},
    moving_variance_initializer= {'class_name': 'Ones', 'config': {}},
    scale= True,
    trainable=True))

model1.add(LeakyReLU(alpha= 0.0,dtype='float32',trainable=True))

model1.add(MaxPooling1D(data_format='channels_last',
    dtype='float32',
    padding='valid',
    pool_size= (1,),
    strides=(3,),
    trainable=True))

model1.add(Flatten(data_format= 'channels_last',
    dtype= 'float32',
  trainable= True))

###
model1.add(Dense(180,activation='sigmoid', activity_regularizer=None,
    bias_constraint= None,
    bias_initializer= {'class_name': 'Zeros', 'config': {}},
    bias_regularizer=None,
    dtype='float32',
    kernel_constraint= {'class_name': 'MaxNorm',
     'config': {'axis': 0, 'max_value': 1}},
    kernel_initializer= {'class_name': 'VarianceScaling',
     'config': {'distribution': 'uniform',
      'mode': 'fan_avg',
      'scale': 1.0,
      'seed': None}},
    kernel_regularizer={'class_name': 'L1L2',
     'config': {'l1': 9.99999993922529e-09, 'l2': 4.999999987376214e-07}}, 
     trainable=True,	
    use_bias= True
     ))

model1.add(Dense(2,activation='softmax',activity_regularizer= None,bias_constraint= None,
                 kernel_constraint= {'class_name': 'MaxNorm',
     'config': {'axis': 0, 'max_value': 1}},
    kernel_initializer= {'class_name': 'VarianceScaling',
     'config': {'distribution': 'uniform',
      'mode': 'fan_avg',
      'scale': 1.0,
      'seed': None}},
    kernel_regularizer= {'class_name': 'L1L2',
     'config': {'l1': 9.99999993922529e-09, 'l2': 4.999999987376214e-07}}
     ))

model1.build()
```

```python
def load_dataset(Dfile):

    print("reading enhancers...")
    data = {}
    with h5py.File(Dfile, "r") as inf:
        for _key in inf:
            data[_key] = inf[_key][()]
    return data
```
